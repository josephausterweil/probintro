## Gaussian 2-Clusters
In this problem, we will investigate how to make categorization decisions for two categories, where each is defined as a Gaussian distribution. First, you will derive the probability of an item being in one category or another. Then, we will explore how the variances and prior probability of each category affect the posterior and predictive distributions.

For this problem, we will assume that data are generated by first picking which of two categories $c=1,2$ it belongs to (according to their prior probability) and then generating the datum according to the corresponding category's likelihood. This results in the following generative process:

$
\begin{align*}
c_n | \theta \sim {\rm Bernoulli}(\theta)   & \qquad \qquad \qquad & x_n | \mu_{c(n)}, \sigma_{c(n)}^2 \overset{iid}{\sim} {\rm N}(\mu_{c(n)}, \sigma_{c(n)}^2)
\end{align*}
$

Note that $c(n)$ is the same as $c_n$. I use parentheses rather than a subscript to avoid having something that is
``double subscripted.''. In generative process notation, $c_n | \theta \sim {\rm Bernoulli}(\theta)$ means that $c_n$, the category for data point $n$, is a Bernoulli random variable with parameter $\theta$. This means $c_n$ will be $1$ with probability $\theta$ (and so with probability $1-\theta$, $c_n=2$). So, the prior probability of category 1 is $\theta$ ($P(c_n=1)=\theta$). For all of Problem 2, assume $\mu_1=-1$ and $\mu_2=1$.

### a. Derivation: Categorization

 Using Bayes' rule, derive the probability of a single datum being in category 1: $P(c_1=1|x_1)$.\footnote{If you are wondering why $P$ is uppercase here and lowercase in other cases, it has to do with whether it is a probability mass (uppercase) or density (lowercase) function. You do not have to worry about this distinction (I mess it up sometimes too!), but as a general rule, you should use the uppercase $P(\cdot)$ when it is a discrete random variable and the lowercase $p(\cdot)$ when it is a continuous random variable.} You can assume that the values of $\mu_1, \mu_2, \sigma^2_1,$ and $\sigma^2_2$ are given parameters (I didn't put them to the right of the $|$ to save space). Show your work (if you do not know how to create equations on a computer, you can scan your handwritten derivation and include it as an image into your document).

As the next problem depends on this answer, I will give you what your derivation should end up with. It is

 $
 \begin{equation*}
P(c=1|x_1) = \frac{\theta {\rm N}(x_1; \mu_1, \sigma_1^2)}{\theta {\rm N}(x_1; \mu_1, \sigma_1^2) + (1-\theta){\rm N}(x_1; \mu_2, \sigma_2^2)}
\end{equation*}
 $

where $N(x;\mu,\sigma^2)$ is the probability density of $x$ from a Normal distribution with mean $\mu$ and variance $\sigma^2$.



#### <font color="blue"> **Derivation**
Using Bayes' rule:

$
\begin{align*}
P(c_1 = 1 | x_1) = \frac{P(x_1 | c_1 = 1) P(c_1 = 1)}{P(x_1)}
\end{align*}
$

1. **Prior probability** of category 1:
  $
  \begin{align*}
  P(c_1 = 1) = \theta
  \end{align*}
  $

2. **Likelihood** of $x_1$ given category 1:
  $
  \begin{align*}
  P(x_1 | c_1 = 1) = \ N(x_1; \mu_1, \sigma_1^2)
  \end{align*}
  $
  where:
  $
  \begin{align*}
  \ N (x; \mu, \sigma^2) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)
  \end{align*}
  $

3. Similarly, for category 2:
   - **Prior probability**:
  $
  \begin{align*}
  P(c_1 = 2) = 1 - \theta
  \end{align*}
  $
     
   - **Likelihood**:
  $
  \begin{align*}
  P(x_1 | c_1 = 2) = \ N(x_1; \mu_2, \sigma_2^2)
  \end{align*}
  $

---

#### **Marginal likelihood $P(x_1)$**

Using the law of total probability:
  $
  \begin{align*}
  P(x_1) = P(x_1 | c_1 = 1) P(c_1 = 1) + P(x_1 | c_1 = 2) P(c_1 = 2)
  \end{align*}
  $


Substitute the terms:
  $
  \begin{align*}
  P(x_1) = \theta \ N(x_1; \mu_1, \sigma_1^2) + (1-\theta) \ N(x_1; \mu_2, \sigma_2^2)
  \end{align*}
  $

---

#### **Posterior probability**

Substitute into Bayes' rule:
$
\begin{align*}
P(c_1 = 1 | x_1) = \frac{P(x_1 | c_1 = 1) P(c_1 = 1)}{P(x_1)}
\end{align*}
$

$
\begin{align*}
P(c_1 = 1 | x_1) = \frac{\theta \ N(x_1; \mu_1, \sigma_1^2)}{\theta \ N(x_1; \mu_1, \sigma_1^2) + (1-\theta) \ N(x_1; \mu_2, \sigma_2^2)}
\end{align*}
$

---

#### <font color="blue"> **Final Answer**


\begin{align*}
P(c_1=1|x_1) = \frac{\theta \ N(x_1; \mu_1, \sigma_1^2)}{\theta \ N(x_1; \mu_1, \sigma_1^2) + (1-\theta) \ N(x_1; \mu_2, \sigma_2^2)}
\end{align*}


where $\ N(x; \mu, \sigma^2)$ is the probability density of $ x$ from a Normal distribution with mean $\mu$ and variance $\sigma^2$.


</font>

<font color="blue"> **Description**: The probability (as $c_1$) of a single datum is the proportion of the likelihood weighted by the prior probability.

Calculate and plot the probability of being in category 1 (so, the x-axis is the $x_1$ value and the y-axis is $P(c_1=1|x_1)$) for $\theta = 0.5$ and for $\theta = 0.75$ with $\sigma_1^2=\sigma_2^2=1$ (remember $\mu_1 = -1$ and $\mu_2=1$).
```python
# dependencies
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# define marginal prob function
def update_datum_c1(mu_1, mu_2, sigma_1_squared, sigma_2_squared, theta, x_min, x_max):
  # Define x range
    x_range = np.linspace(x_min, x_max, 1000)

    # Compute likelihoods
    likelihood_1 = norm.pdf(x_range, mu_1, np.sqrt(sigma_1_squared))
    likelihood_2 = norm.pdf(x_range, mu_2, np.sqrt(sigma_2_squared))

    # Compute posterior probabilities
    posterior_c1 = (theta * likelihood_1) / (
        theta * likelihood_1 + (1 - theta) * likelihood_2
    )

    # Compute marginal distribution
    marginal = theta * likelihood_1 + (1 - theta) * likelihood_2

    return posterior_c1, marginal
```

```python
# fill me
# According to the result from last question, we could express the P(c_1|x_1)
# By multiplying theta_i and N(x_1; miu_i, sigma_i^2), and sum up as the denominator

# Parameters
theta_values = [0.5, 0.75, 0.25]  # prior probabilities for category 1
mu_1, mu_2 = -1, 1  # means for categories 1 and 2

# Variance configurations
configs = [
    (1, 1),  # _1^2 = _2^2 = 1
    (0.5, 2),  # _1^2 = 0.5, _2^2 = 2
    (2, 0.5)
]

# Colors for different theta values
colors = {0.5: "blue", 0.75: "orange", 0.25: "#980025"}
colors_sigma = {1: "steelblue", 0.5: "red", 2: "green"}
linestyles = {1: "-", 0.5: "--", 2: ":"}

# Canvas for plots (two subplots: one for posterior, one for predictive)
fig, axes = plt.subplots(2, 1, figsize=(12, 10))
x_min = -10
x_max = 10
x_range = np.linspace(x_min, x_max, 1000)

for j, (sigma_1_squared, sigma_2_squared) in enumerate(configs):
    for theta in theta_values:
        posterior_c1, _ = update_datum_c1(mu_1, mu_2, sigma_1_squared, sigma_2_squared, theta, x_min, x_max)

        # Plot posterior probability for P(c1=1 | x)
        axes[0].plot(
            x_range,
            posterior_c1,
            label=f"$P(c_1=1|x_1), \\theta={theta}, \\sigma_1^2={sigma_1_squared}, \\sigma_2^2={sigma_2_squared}$",
            color=colors[theta],
            linestyle=linestyles[sigma_1_squared],
        )

for sigma_1_squared, sigma_2_squared in configs:
    axes[1].plot(
        x_range,
        norm.pdf(x_range, mu_1, np.sqrt(sigma_1_squared)),
        label=f"$L(c_1=1|x_1), \\sigma_1^2={sigma_1_squared}, \\sigma_2^2={sigma_2_squared}$",
        color=colors_sigma[sigma_1_squared],
        linestyle="-"
    )

    axes[1].plot(
        x_range,
        norm.pdf(x_range, mu_2, np.sqrt(sigma_2_squared)),
        label=f"$L(c_2=1|x_1), \\sigma_1^2={sigma_1_squared}, \\sigma_2^2={sigma_2_squared}$",
        color=colors_sigma[sigma_1_squared],
        linestyle="--"
    )

# Configure posterior plot
axes[0].set_title("Fig 1. Posterior Probability $P(c_1=1|x)$")
axes[0].set_xlabel("$x$")
axes[0].set_ylabel("Probability")
axes[0].legend()
axes[0].grid()

# likelihoods
axes[1].set_title("Fig 2. Likelihoods")
axes[1].set_xlabel("$x$")
axes[1].set_ylabel("Density")
axes[1].legend()
axes[1].grid()

# Adjust layout and display
plt.tight_layout()
plt.show()
```

### c. Derivation --- Prediction

Using Bayes' rule and the {\em Law of Total Probability}, derive the probability of a data point $p(x)$ according to this model (note, this is without any given data). As the next problem depends on this answer, I will give you what your derivation should end up with. It is

#### fill me

#### <font color="blue"> **Derivation**
Using the law of total probability, the marginal likelihood \( P(x_1) \) can be derived as follows:

$
\begin{align*}
P(x_1) &= \sum_{c} P(x_1 | c) P(c)
\end{align*}
$

---

#### <font color="blue"> **Step 1: Expand for two categories**

For $c = 1$ and $c = 2$:
$
\begin{align*}
P(x_1) &= P(x_1 | c_1 = 1) P(c_1 = 1) + P(x_1 | c_1 = 2) P(c_1 = 2)
\end{align*}
$

---

#### <font color="blue"> **Step 2: Substitute terms**

1. **Prior probabilities**:
   $
   \begin{align*}
   P(c_1 = 1) &= \theta, \\
   P(c_1 = 2) &= 1 - \theta
   \end{align*}
   $

2. **Likelihoods**:
   $
   \begin{align*}
   P(x_1 | c_1 = 1) &= \ N(x_1; \mu_1, \sigma_1^2), \\
   P(x_1 | c_1 = 2) &= \ N(x_1; \mu_2, \sigma_2^2)
   \end{align*}
   $

3. **Gaussian distribution**:
   $
   \begin{align*}
   \ N(x; \mu, \sigma^2) &= \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)
   \end{align*}
   $

---

#### <font color="blue"> **Step 3: Combine terms**

Substituting into the marginal likelihood:
$
\begin{align*}
P(x_1) &= \theta \cdot \ N(x_1; \mu_1, \sigma_1^2) + (1-\theta) \cdot \ N(x_1; \mu_2, \sigma_2^2)
\end{align*}
$

---

#### <font color="blue"> **Final Answer**

$
\begin{align*}
P(x_1) &= \theta \cdot \ N(x_1; \mu_1, \sigma_1^2) + (1-\theta) \cdot \ N(x_1; \mu_2, \sigma_2^2)
\end{align*}
$
